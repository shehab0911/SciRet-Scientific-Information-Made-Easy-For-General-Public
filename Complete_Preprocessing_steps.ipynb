```python
#Extract only introduction text from body_text 
import pandas as pd

# load the CSV file into a Pandas DataFrame
df = pd.read_csv("/content/drive/MyDrive/RAG AND GPT/new_dataframe (1).csv")

# define a function to extract introduction text from a string
def extract_intro(text):
    # split the text into lines
    lines = text.split('\n')

    # initialize an empty string to store the introduction
    intro = ''

    # loop over each line
    for line in lines:
        # check if the line contains the word "Introduction"
        if "Introduction" in line:
            # concatenate the line to the introduction string
            intro += line + '\n'

    # return the introduction text
    return intro

# apply the extract_intro function to the body_text column
df['introduction'] = df['body_text'].apply(extract_intro)

# print the introduction text for each row
for intro in df['introduction']:
    print(intro[:100])
```

```python
import pandas as pd
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob

# load the CSV file into a pandas dataframe
df = pd.read_csv('filename.csv')

# rename the metadata column to title
df = df.rename(columns={'metadata': 'title'})

# extract the introduction from the body text
df['introduction'] = df['body_text'].str.split('\n\n', 1).str[0]

# drop the original body_text column
df.drop('body_text', axis=1, inplace=True)

# tokenize the title and introduction columns
df['title'] = df['title'].apply(nltk.word_tokenize)
df['introduction'] = df['introduction'].apply(nltk.word_tokenize)

# convert words to lowercase in the title and introduction columns
df['title'] = df['title'].apply(lambda x: [word.lower() for word in x])
df['introduction'] = df['introduction'].apply(lambda x: [word.lower() for word in x])

# apply spell correction to the title and introduction columns
df['title'] = df['title'].apply(lambda x: [str(TextBlob(word).correct()) for word in x])
df['introduction'] = df['introduction'].apply(lambda x: [str(TextBlob(word).correct()) for word in x])

# apply lemmatization to the title and introduction columns
lemmatizer = WordNetLemmatizer()
df['title'] = df['title'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
df['introduction'] = df['introduction'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

# remove punctuation from the title and introduction columns
df['title'] = df['title'].apply(lambda x: [word for word in x if word not in string.punctuation])
df['introduction'] = df['introduction'].apply(lambda x: [word for word in x if word not in string.punctuation])

# remove stop words from the title and introduction columns
stop_words = set(stopwords.words('english'))
df['title'] = df['title'].apply(lambda x: [word for word in x if word not in stop_words])
df['introduction'] = df['introduction'].apply(lambda x: [word for word in x if word not in stop_words])

# apply parts of speech tagging to the title and introduction columns
df['title_pos'] = df['title'].apply(nltk.pos_tag)
df['introduction_pos'] = df['introduction'].apply(nltk.pos_tag)

# save the updated dataframe to a new CSV file with only title and introduction columns
df[['title', 'title_pos', 'introduction', 'introduction_pos']].to_csv('new_filename.csv', index=False)

```

**Removing URLs**

```python
import re
df['title'] = df['title'].apply(lambda x: [re.sub(r'http\S+', '', word) for word in x])
df['introduction'] = df['introduction'].apply(lambda x: [re.sub(r'http\S+', '', word) for word in x])
```
